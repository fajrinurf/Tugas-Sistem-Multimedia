# Point Operation

```python
def point_operation(image, method):
    for row in range(0, image.shape[0]):
        for col in range(0, image.shape[1]):
            image[row][col] = method(image[row][col])
    return image
```

### Contrast Modification

```python
def f(x):
    return x * 1.5
```

```python
contrast_modified = point_operation(gray1.copy(), f)
```

```python
imshow(contrast_modified)
```

```python
plt.hist(contrast_modified.ravel(), bins=256, histtype='step', color='black')
plt.show()
```

### Inverting an Image

```python
def point_operation(image):
    a_max = image.max()
    for row in range(0, image.shape[0]):
        for col in range(0, image.shape[1]):
            image[row][col] = (image[row][col] * -1) + a_max
    return image
```

```python
image_invert = point_operation(gray1.copy())
```

```python
imshow(image_invert)
```

```python
plt.hist(image_invert.ravel(), bins=256, histtype='step', color='black')
plt.show()
```

### Thresholding an Image

```python
def point_operation(image, threshold):
    for row in range(0, image.shape[0]):
        for col in range(0, image.shape[1]):
            if image[row][col] >= threshold:
               image[row][col] = 255
            else:
               image[row][col] = 0
    return image
```

```python
thresholding = point_operation(gray1.copy(), 150)
```

```python
imshow(thresholding)
```

```python
plt.hist(thresholding.ravel(), bins=256, histtype='step', color='black')
plt.show()
```
