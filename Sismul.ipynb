# Perbandingan Citra Asli dengan Equalisation dan Streching

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 5))

ax1.set_title('Citra Asli')
ax1.imshow(gray1)

ax2.set_title('Histogram Equalisation')
ax2.imshow(gray2)

ax3.set_title('Histogram Streching')
ax3.imshow(gray3)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 5))

ax1.set_title('Citra Asli')
ax1.hist(gray1.ravel(), bins=256, histtype='step', color='black')

ax2.set_title('Histogram Equalisation')
ax2.hist(img_as_ubyte(img_equ).ravel(), bins=256, histtype='step', color='black')

ax3.set_title('Histogram Streching')
ax3.hist(img_as_ubyte(gray3).ravel(), bins=256, histtype='step', color='black')

plt.show()
```
